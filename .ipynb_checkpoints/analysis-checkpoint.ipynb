```python
import re
re.findall("(\d+) (mah)", "078 mah", re.IGNORECASE)
```

```python
def get_type(row):
    wrong_col = row.index[(row.str.contains("Li-") == True) | (row.str.contains("removable", case=False) == True) | 
                          (row.str.contains("(\d+) (mah)", case=False) == True)]
    if len(wrong_col):
        row["Type"] = row[wrong_col].values[0]
        row[wrong_col] = np.nan
    return row
```

```python
%%time
extracted_df["Battery"] = extracted_df["Battery"].apply(get_type, axis=1)
```

```python
%%time
def get_charging(row):
    wrong_col = row.drop("Charging").index[(row.drop("Charging").str.contains("charging", case=False) == True) |
                                           (row.drop("Charging").str.contains("quick charge", case=False) == True)]
    if len(wrong_col):
        row["Charging"] = row[wrong_col].values[0].strip()
        row[wrong_col] = np.nan
    return row
extracted_df["Battery"] = extracted_df["Battery"].apply(get_charging, axis=1)
```
