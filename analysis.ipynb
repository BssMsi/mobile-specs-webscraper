```python
def extract_specs(row):
    global specifications
    specifications = specifications.append(pd.DataFrame(
    pd.concat([pd.DataFrame(i) for i in literal_eval(row)], axis=1).stack()).T.swaplevel(0, 1, axis=1), 
                         ignore_index=True)
    return row
```

```python
df = pd.read_csv("gsmarena_data.csv", header=0)
```

```python
%%time
specifications = pd.DataFrame()
df['Specifications'].apply(extract_specs)
```

```python
%%time
df.columns = pd.MultiIndex.from_tuples([("Brand", "Name"), ("Model", "Name"), ("Specs", "data")])
extracted_df = pd.concat([df.drop('Specs', axis=1), specifications], axis=1)
```

```python
extracted_df.to_csv("specs_extracted.csv", index=False)
```
